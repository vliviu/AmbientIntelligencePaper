```python
#  https://towardsdatascience.com/understanding-the-confusion-matrix-from-scikit-learn-c51d88929c79
import matplotlib.pyplot as plt
import numpy as np
import seaborn
import csv
from sklearn.metrics import confusion_matrix
```

```python
actual = []
with open('actual.csv', mode ='r') as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC) # change contents to floats
    for row in reader: # each row is a list
        actual.append(row)
```

```python
predicted = []
with open('predicted.csv', mode ='r') as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC) # change contents to floats
    for row in reader: # each row is a list
        predicted.append(row)
```

```python
def plot_confusion_matrix(data, labels, output_filename):
    """Plot confusion matrix using heatmap.
 
    Args:
        data (list of list): List of lists with confusion matrix data.
        labels (list): Labels which will be plotted across x and y axis.
        output_filename (str): Path to output file.
 
    """
    seaborn.set(color_codes=True)
    plt.figure(1, figsize=(18, 15))
 
    plt.title("Confusion Matrix")
 
    seaborn.set(font_scale=0.85)
    ax = seaborn.heatmap(data, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Scale'})
 
    ax.set_xticklabels(labels)
    ax.set_yticklabels(labels)
 
    ax.set(ylabel="True Label", xlabel="Predicted Label")
 
    plt.savefig(output_filename, bbox_inches='tight', dpi=600)
    plt.show()
    plt.close()
```

```python
# define labels
labels = ["RH-D_LH-D","RH-L_LH-D","RH-F_LH-D","RH-U_LH-D","RH-D_LH-U","RH-L_LH-U","RH-L_LH-L","RH-F_LH-F","RH-U_LH-F","RH-U_LH-U"]
 
# create confusion matrix
plot_confusion_matrix(predicted, labels, "confusion_matrix.png")
```
